```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image

# Abre la imagen
img = Image.open("/kaggle/input/publicsargazods/images/k0006.jpg")

# Especifica las coordenadas del área que se desea recortar (izquierda, arriba, derecha, abajo)
area = (100, 100, 300, 300)

# Recorta la imagen utilizando la función crop
subventana = img.crop(area)

# Cierra la imagen original
img.close()

plt.imshow(subventana)

plt.show()
```

Conversión a escala de grises de la subventana

```python
# Convierte la imagen a modo RGB si no lo está
if subventana.mode != "RGB":
    subventana = subventana.convert("RGB")

# Aplica la conversión a escala de grises utilizando la fórmula NTSC
imagen_escala_grises = subventana.copy()
pixels = imagen_escala_grises.load()
width, height = imagen_escala_grises.size

for i in range(width):
    for j in range(height):
        r, g, b = imagen_escala_grises.getpixel((i, j))
        gray = int(0.299 * r + 0.587 * g + 0.114 * b)
        pixels[i, j] = (gray, gray, gray)

plt.imshow(imagen_escala_grises)

plt.show()

# Cierra la imagen recortada
subventana.close()
```
